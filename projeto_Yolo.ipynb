# **Referencial Teórico**
##*YOLOv8* 
O algoritmo para a detecção das frutas que será utilizado é o YOLOv8, ele é o mais recentes da família YOLO, desenvolvido pela Ultralytics, YOLOv8 foi criado pra ser rápido, preciso e facil de manejar, esse algoritmo é muito utilizado em sistemas de monitoramento em tempo real, como câmeras de CFTV e veículos autônomos.

image.png

O YOLO funciona inicialmente dividindo a imagem em um grid de células. Cada célula é responsável por prever um conjunto de caixas delimitadoras que podem conter objetos. Em seguida, são realizadas a regressão das coordenadas das caixas, atribuição de pontuações e classificação dos objetos contidos nessas caixas.

## Roboflow
É uma plataforma de computação visual que permite ao usuário construir rapidamente modelos mais precisos a partir de coleções de dados, pré-processamento e técnicas de treinamento de modelo. Ela permite que o usuário faça upload imagens e vídeos personalizados, realize anotações nessas mídias para rotular os objetos de interesse, modifique a orientação e o tamanho da imagem e treine os modelos de machine learning com mais facilidade.
